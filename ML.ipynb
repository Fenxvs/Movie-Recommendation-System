```python
import pandas as pd
```

```python
df = pd.read_csv("cleaned_movie_dataset.csv")
df.head()
```

```python
movies=df[['id', 'title', 'overview', 'genre']]
movies.head()
```

```python
movies['tags'] = movies['overview']+movies['genre']
movies.head()
```

```python
new_movies  = movies.drop(columns=['overview', 'genre'])
new_movies.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import pickle
```

```python
cv=CountVectorizer(max_features=10000, stop_words='english')
cv
```

```python
vector=cv.fit_transform(new_movies['tags'].values.astype('U')).toarray()
vector.shape
```

```python
similarity=cosine_similarity(vector)
similarity
```

```python
new_movies[new_movies['title']=="The Godfather"].index[0]
```

```python
distance = sorted(list(enumerate(similarity[2])), reverse=True, key=lambda vector:vector[1])
for i in distance[0:5]:
    print(new_movies.iloc[i[0]].title)
```

```python
def recommand_movie(movies):
    index=new_movies[new_movies['title']==movies].index[0]
    distance = sorted(list(enumerate(similarity[index])), reverse=True, key=lambda vector:vector[1])
    for i in distance[0:5]:
        print(new_movies.iloc[i[0]].title)
```

```python
recommand_movie("Iron Man")
```

```python
pickle.dump(new_movies, open('movies_list.pkl', 'wb'))
```
